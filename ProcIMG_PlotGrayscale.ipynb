```python
import skimage as ski
import matplotlib.pyplot as plt
import numpy as np
import os
from skimage import color
from skimage import io
import matplotlib.pyplot as plt
import math

for filename in os.listdir('./images'):
        filepath = os.path.join('./images', filename)
        if os.path.isfile(filepath):
            print(f"Reading file: {filepath}")
            image = ski.io.imread(filepath)
            grayscale_img = color.rgb2gray(image)
            plt.title(f"Pixel (10,10) - {(grayscale_img[10][10] * 255):.2f} ({math.floor(grayscale_img[10][10]*100)}%)")
            plt.imshow(grayscale_img, cmap='gray')
            plt.show()


# coins_gray = ski.color.rgb2gray(coins)

# plt.imshow(coins_gray, cmap='gray')
# plt.show()
```
